```python
# resizing image using opencv - upscaling

import cv2 as cv
import matplotlib.pyplot as plt


img = cv.imread(r'/content/ant.jpg',1)
scale = 150 # upscaling

height = int(img.shape[0] * scale / 100)
width = int(img.shape[1] * scale / 100)

dim = (width,height)

resized = cv.resize(img,dim)
resized = cv.cvtColor(resized,cv.COLOR_BGR2RGB)
plt.imshow(resized)

print(img.shape)
print(resized.shape)



```

```python
# Roatating of image

import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread(r'/content/ant.jpg',1)


# get rows and column of the image

ht,wd = img.shape[:2]

#300,300 - CP 150,150
#create a rotation matix M

M = cv.getRotationMatrix2D((ht/2,wd/2),90,0.75) # 1 means retain original image size

# rotate image with rotation matrix
dst = cv.warpAffine(img,M,(ht,wd))

# rotate the image
dst = cv.cvtColor(dst,cv.COLOR_BGR2RGB)

# SHOW THE ROTATED IMAGE

plt.imshow(dst)

```

```python

 # Logical Operation On Image
 
import cv2 as cv
import matplotlib.pyplot as plt


image1 = cv.imread(r'/Image 1.jpg',1)
image2 = cv.imread(r'/img 11.jpg',1)

plt.figure(figsize=(10,7),facecolor = 'lime')
plt.subplot(2,3,1)
img1 = cv.cvtColor(image1,cv.COLOR_BGR2RGB)
plt.title('Original Image 1')
plt.imshow(img1)

plt.subplot(2,3,2)
img2 = cv.cvtColor(image2,cv.COLOR_BGR2RGB)
plt.title('Original Image 2')
plt.imshow(img2)

plt.subplot(2,3,3)
img3 = cv.bitwise_or(image1,image2)
plt.title('Bitwise_OR Img')
img3 = cv.cvtColor(img3,cv.COLOR_BGR2RGB)
plt.imshow(img3)

plt.subplot(2,3,4)
img4 = cv.bitwise_and(image1,image2)
img4 = cv.cvtColor(img4,cv.COLOR_BGR2RGB)
plt.title('Bitwise_And Img')
plt.imshow(img4)

plt.subplot(2,3,5)
img5 = cv.bitwise_xor(image1,image2)
img5 = cv.cvtColor(img5,cv.COLOR_BGR2RGB)
plt.title('Bitwise_XOR Img')
plt.imshow(img5)

plt.subplot(2,3,6)
img6 = cv.bitwise_not(image1,image2)
img6 = cv.cvtColor(img6,cv.COLOR_BGR2RGB)
plt.title('Bitwise_NOT Img')
plt.imshow(img6)


```

```python
# image thresholding using simple threshoulding

import cv2 as cv
import matplotlib.pyplot as plt

image1 = cv.imread(r'/content/ant.jpg')

_,thresh_binary = cv.threshold(image1,120,255,cv.THRESH_BINARY)

# if pixel value > threshold - pixel value = 255 else pixel = 0
plt.figure(figsize=(10,5),facecolor="yellow")
plt.suptitle("Image Threshoulding using simple Threshoulding")
plt.subplot(1,2,1)
image1 = cv.cvtColor(image1,cv.COLOR_BGR2RGB)
plt.imshow(image1)

plt.subplot(1,2,2)
plt.imshow(thresh_binary,'gray')
```

```python
# Image thresholding using simple Thresholding

import cv2 as cv
import matplotlib.pyplot as plt

image1 = cv.imread(r'/content/ant.jpg',1)

# if pixel value > threshoulding - pixel value = 255 else pixel = 0
plt.figure(figsize=(10,7),facecolor='olivedrab')
plt.suptitle("Image Threshoulding using simle Threshoulding")
plt.subplot(2,3,1)
image1 = cv.cvtColor(image1,cv.COLOR_BGR2RGB)
plt.title("Original Image")
plt.imshow(image1)

plt.subplot(2,3,2)
ret,thresh_binary = cv.threshold(image1,120,255,cv.THRESH_BINARY)
plt.title("Thresh_Binary Image")
print(ret)
plt.imshow(thresh_binary,'gray')

#if pixel value > threshold - pixel value - 0 else pixel = 255

plt.subplot(2,3,3)
ret,thresh_binary_inv = cv.threshold(image1,120,255,cv.THRESH_BINARY_INV)
plt.title("Thresh_Binary_Inv Image")
print(ret)
plt.imshow(thresh_binary_inv,'gray')

plt.subplot(2,3,4)
ret,thresh_trunc = cv.threshold(image1,120,255,cv.THRESH_TRUNC)
plt.title("Thresh_Trunc Image")
print(ret)
plt.imshow(thresh_trunc,'gray')


plt.subplot(2,3,5)
ret,thresh_tozero = cv.threshold(image1,120,255,cv.THRESH_TOZERO)
plt.title("Thresh_Tozero Image")
print(ret)
plt.imshow(thresh_tozero,'gray')

plt.subplot(2,3,6)
ret,thresh_tozero_inv = cv.threshold(image1,120,255,cv.THRESH_TOZERO_INV)
plt.title("Thresh_Tozero_Inv Image")
print(ret)
plt.imshow(thresh_tozero_inv,'gray')

```

```python
# Image Thresholding using open cv

import cv2 as cv
import matplotlib.pyplot as plt

image1 = cv.imread(r'/content/ant.jpg',0)


plt.figure(figsize=(10,7),facecolor='limegreen')
ret,thresh_binary = cv.threshold(image1,150,255,cv.THRESH_BINARY)
ret,thresh_binary_inv = cv.threshold(image1,150,255,cv.THRESH_BINARY_INV)
ret,thresh_trunc = cv.threshold(image1,150,255,cv.THRESH_TRUNC)
ret,thresh_tozero = cv.threshold(image1,150,255,cv.THRESH_TOZERO)
ret,thresh_tozero_inv = cv.threshold(image1,150,255,cv.THRESH_TOZERO_INV)

names = ['ORIGINAL IMAGE','BINARY','THRESH_BINARY_INV','THRESH_TRUNC',\
         'THRESH_TOZERO','THRESH_TOZERO_INV']
    
images = [image1,thresh_binary,thresh_binary_inv,thresh_trunc,\
          thresh_tozero,thresh_tozero_inv]

for i in range(6):
    plt.subplot(2,3,i+1)
    plt.suptitle('Image Threshoulding using Open CV')
    plt.imshow(images[i],'gray')
    plt.title(names[i])
    plt.xticks([]),plt.yticks([])
    
plt.show()    
```
